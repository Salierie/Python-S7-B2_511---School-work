## Tuple

1. to create a tuple of numbers and print one item.
2. to unpack a tuple into several variables.
3. to add an item to a tuple.
4. to find the index of an item in a tuple.
5. to find the repeated items of a tuple

```python
# Create a tuple of numbers and print one item
numbers = (1, 2, 3, 4, 5)
print(numbers[2])  # Prints 3

# Unpack a tuple into several variables
a, b, c, d, e = numbers
print(a, b, c, d, e)  # Prints 1 2 3 4 5

# Add an item to a tuple (tuples are immutable, so create a new tuple)
new_numbers = numbers + (6,)
print(new_numbers)  # Prints (1, 2, 3, 4, 5, 6)

# Find the index of an item in a tuple
index = numbers.index(3)
print(index)  # Prints 2

# Find the repeated items of a tuple
def find_repeated_items(tup):
    repeated = set()
    for item in tup:
        if tup.count(item) > 1:
            repeated.add(item)
    return repeated

example_tuple = (1, 2, 2, 3, 4, 4, 4, 5)
print(find_repeated_items(example_tuple))  # Prints {2, 4}
```

## Set
1. Write a Python program to find the maximum and minimum values in a set.
2. Write a Python program to check if a given value is present in a set or not.
3. Write a Python program to check if two given sets have no elements in common.
4. Write a Python program to find all the unique words and count the frequency of occurrence from a given list of strings. Use Python set data type.
5. Given two sets of numbers, write a Python program to find the missing numbers in the second set as compared to the first and vice versa. Use the Python set.

```python
# Find the maximum and minimum values in a set
my_set = {1, 2, 3, 4, 5}
print(max(my_set))  # Prints 5
print(min(my_set))  # Prints 1

# Check if a given value is present in a set or not
value = 3
if value in my_set:
    print(f"{value} is in the set")
else:
    print(f"{value} is not in the set")

# Check if two given sets have no elements in common
set1 = {1, 2, 3}
set2 = {4, 5, 6}
if set1.isdisjoint(set2):
    print("No common elements")
else:
    print("Have common elements")

# Find all the unique words and count the frequency of occurrence from a given list of strings
def unique_words_and_frequency(strings):
    words = ' '.join(strings).split()
    unique_words = set(words)
    frequency = {word: words.count(word) for word in unique_words}
    return frequency

strings = ["hello world", "world of python", "python is great"]
print(unique_words_and_frequency(strings))

# Find the missing numbers in the second set as compared to the first and vice versa
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
missing_in_b = set_a - set_b
missing_in_a = set_b - set_a
print(f"Missing in B: {missing_in_b}")
print(f"Missing in A: {missing_in_a}")
```

## Dict
1. Convert two lists into a dictionary
2. Merge two Python dictionaries into one
3. Print the value of key ‘history’ from the below dict
4. Initialize dictionary with default values
5. Create a dictionary by extracting the keys from a given dictionary
6. Delete a list of keys from a dictionary
7. Check if a value exists in a dictionary
8. Rename key of a dictionary
9. Get the key of a minimum value from the following dictionary
10. Change value of a key in a nested dictionary
11. Counts the number of times characters appear in a text paragraph.
12. Using a dictionary containing keys starting from 1 and values ​​containing prime numbers less than a value N.

```python
# Convert two lists into a dictionary
keys = ['a', 'b', 'c']
values = [1, 2, 3]
my_dict = dict(zip(keys, values))
print(my_dict)  # Prints {'a': 1, 'b': 2, 'c': 3}

# Merge two Python dictionaries into one
dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}
merged_dict = {**dict1, **dict2}
print(merged_dict)  # Prints {'a': 1, 'b': 2, 'c': 3, 'd': 4}

# Print the value of key 'history' from the dictionary
sample_dict = {'class': {'student': {'name': 'Mike', 'marks': {'physics': 70, 'history': 80}}}}
print(sample_dict['class']['student']['marks']['history'])  # Prints 80

# Initialize dictionary with default values
keys = ['a', 'b', 'c']
default_value = 0
default_dict = dict.fromkeys(keys, default_value)
print(default_dict)  # Prints {'a': 0, 'b': 0, 'c': 0}

# Create a dictionary by extracting the keys from a given dictionary
given_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
keys_to_extract = ['a', 'c']
extracted_dict = {key: given_dict[key] for key in keys_to_extract}
print(extracted_dict)  # Prints {'a': 1, 'c': 3}

# Delete a list of keys from a dictionary
keys_to_delete = ['b', 'd']
for key in keys_to_delete:
    given_dict.pop(key, None)
print(given_dict)  # Prints {'a': 1, 'c': 3}

# Check if a value exists in a dictionary
value_to_check = 3
if value_to_check in given_dict.values():
    print(f"{value_to_check} exists")
else:
    print(f"{value_to_check} does not exist")

# Rename key of a dictionary
old_key = 'a'
new_key = 'alpha'
if old_key in given_dict:
    given_dict[new_key] = given_dict.pop(old_key)
print(given_dict)  # Prints {'alpha': 1, 'c': 3}

# Get the key of a minimum value from the dictionary
sample_dict = {'Physics': 82, 'Math': 65, 'History': 75}
min_key = min(sample_dict, key=sample_dict.get)
print(min_key)  # Prints 'Math'

# Change value of a key in a nested dictionary
nested_dict = {'person': {'name': 'Alice', 'age': 30}}
nested_dict['person']['age'] = 31
print(nested_dict)  # Prints {'person': {'name': 'Alice', 'age': 31}}

# Count the number of times characters appear in a text paragraph
def char_frequency(paragraph):
    frequency = {}
    for char in paragraph:
        if char in frequency:
            frequency[char] += 1
        else:
            frequency[char] = 1
    return frequency

text = "hello world"
print(char_frequency(text))

# Dictionary with keys starting from 1 and values containing prime numbers less than N
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def primes_less_than_N(N):
    primes = [n for n in range(2, N) if is_prime(n)]
    return {i+1: primes[i] for i in range(len(primes))}

N = 20
print(primes_less_than_N(N))

# Restructure company data by grouping employees by department
def group_employees_by_department(employees):
    dept_employees = {}
    for emp_id, info in employees.items():
        dept = info['department']
        if dept not in dept_employees:
            dept_employees[dept] = []
        dept_employees[dept].append({'name': info['name'], 'salary': info['salary']})
    return dept_employees

employees = {
    1: {'name': 'Alice', 'department': 'HR', 'salary': 50000},
    2: {'name': 'Bob', 'department': 'Engineering', 'salary': 60000},
    3: {'name': 'Charlie', 'department': 'HR', 'salary': 55000},
    4: {'name': 'David', 'department': 'Engineering', 'salary': 65000}
}

print(group_employees_by_department(employees))
```
