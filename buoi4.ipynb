# Buoi 4 - 1

```python
"""
Viết 1 game chgo phép người dùng đoán số từ 1 đến 100.
Game có 3 cấp độ chơi:
    - dễ - đoán tối đa 9 lần
    - vừa - đoán tối đa 6 lần
    - khó - đoán tối đa 4 lần
Sau khi người dùng hoàn tất 1 lần chơi,
    chương trình sẽ hỏi người dùng có chơi nữa không.
    - Nếu người chơi đồng ý thì tiếp tục 1 lần chơi mới.
    - Nếu không thì kết thúc trò chơi, thống kê số lần chơi thắng/thua
"""

import random

def play_game():
    levels = {"easy": 9, "medium": 6, "hard": 4}
    
    level = input("Chọn cấp độ (easy, medium, hard): ").lower()
    if level not in levels:
        print("Cấp độ không hợp lệ. Kết thúc trò chơi.")
        return False
    
    number_to_guess = random.randint(1, 100)
    attempts_left = levels[level]
    
    print(f"Bạn có {attempts_left} lần đoán!")
    
    while attempts_left > 0:
        guess = int(input(f"Nhập số bạn đoán (1-100): "))
        
        if guess < number_to_guess:
            print("Số bạn đoán nhỏ hơn số cần tìm.")
        elif guess > number_to_guess:
            print("Số bạn đoán lớn hơn số cần tìm.")
        else:
            print(f"Chúc mừng! Bạn đã đoán đúng số {number_to_guess}!")
            return True  
    
        attempts_left -= 1
        print(f"Bạn còn {attempts_left} lần đoán nữa.")
    
    print(f"Bạn đã hết lượt đoán. Số cần tìm là {number_to_guess}.")
    return False  

def main():
    play_again = True
    win_count = 0
    lose_count = 0
    
    while play_again:
        if play_game():
            win_count += 1
        else:
            lose_count += 1
        
        play_again_input = input("Bạn có muốn chơi lại không? (yes/no): ").lower()
        if play_again_input != "yes":
            play_again = False
    
    print(f"\nKết quả cuối cùng:")
    print(f"Bạn đã thắng {win_count} lần.")
    print(f"Bạn đã thua {lose_count} lần.")

main()
```

# Buoi 4 - 2

```python
"""
Viết 1 game Tài xỉu dưa trên nguyên tắc gieo 2 con xúc sắc ngẫu nhiên.
Nếu tổng giá trị của 2 con xúc sắc > 5, gọi là "Tài"
Còn không, gọi là "Xỉu"

Cho người dùng chọn 1 trong 2 trạng thái Tài hoặc Xỉu.
So sánh kết quả.
Hỏi người chơi có tiếp tục không? Nếu có thì chơi tiếp cho dến khi người dùng chọn không.

Mở rộng:
    1. Bắt đầu chơi, cho người dùng 1 số tiền là 100,000. Số tiền cược của mỗi lần chơi
        là 10,000. Sau khi người dùng chọn không chơi tiếp hoặc hết tiền thì ngưng. Sau đó thống kê
        tiền của người chơi, số ván thắng. (có thể cho người dùng chọn số tiền đặt cược)
    2. Cho phép người dùng cược vào số 5 (3 trạng thái thay vì 2 (Tài/Xỉu). Nếu người dùng cược đúng
        thì thắng được 3 lần số tiền cược.
"""

import random

def roll_dice():
    return random.randint(1, 6), random.randint(1, 6)

def play_game(balance, bet_amount):
    print(f"Bạn hiện có {balance} tiền.")
    print(f"Số tiền cược cho mỗi ván là {bet_amount}.")

    bet_choice = input("Chọn cược (Tài, Xỉu, 5): ").lower()
    while bet_choice not in ["tài", "xỉu", "5"]:
        bet_choice = input("Câu trả lời không hợp lệ, chọn lại (Tài, Xỉu, 5): ").lower()

    dice1, dice2 = roll_dice()
    total = dice1 + dice2
    print(f"Bạn đã lăn được: {dice1} và {dice2}, tổng là {total}")

    if total > 5:
        result = "tài"
    elif total < 5:
        result = "xỉu"
    else:
        result = "5"
    
    print(f"Kết quả là: {result}")

    if bet_choice == result:
        print("Chúc mừng, bạn thắng!")
        balance += bet_amount if bet_choice != "5" else bet_amount * 3
        return True, balance
    else:
        print("Rất tiếc, bạn thua!")
        balance -= bet_amount
        return False, balance

def main():
    balance = 100000  
    bet_amount = 10000  
    
    win_count = 0  

    while balance >= bet_amount:
        print(f"\nSố tiền còn lại: {balance}")
        
        # Chơi 1 ván
        win, balance = play_game(balance, bet_amount)
        
        if win:
            win_count += 1

        if balance < bet_amount:
            print("Bạn không còn đủ tiền để tiếp tục chơi. Kết thúc trò chơi.")
            break

        play_again = input("Bạn có muốn chơi tiếp không? (y/n): ").lower()
        if play_again != "y":
            break
    
    print(f"\nKết quả cuối cùng:")
    print(f"Số tiền còn lại: {balance}")
    print(f"Số ván thắng: {win_count}")

# Chạy trò chơi
main()
```
